```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Obtenemos el dataset de entrenamiento
df_train = pd.read_csv('train.csv').iloc[:, 1:]
df_train
```

```python
df_train.describe()
```

PREPROCESADO

```python
#Buscar si hay filas duplicadas y eliminarlas
df_train.drop_duplicates()
```

```python
df_train=df_train.dropna()
```

```python
df_train.corr()
#vemos que los mas adecuados son fruitset,fruitmass,seeds
```

```python
#Aislamos la columna yield que vamos a predecir
df_yield = df_train['yield'];
df_train = df_train[['fruitset','fruitmass','seeds']];
df_train
```

```python
df_train.corr()
```

MODELO DE REGRESIÃ“N: LinearRegression

```python
X_regression = df_train #Datos que usamos para entrenar al modelo
Y_regression = df_yield #Lo que queremos predecir
```

```python
from sklearn.ensemble import HistGradientBoostingRegressor

modelregresion = HistGradientBoostingRegressor()
modelregresion.fit(pd.DataFrame(X_regression), Y_regression)
```

```python
#Hacemos las predicciones con los datos que usamos para entrenar al modelo
y_pred_regresion = modelregresion.predict(pd.DataFrame(X_regression))
```

```python
df_test = pd.read_csv('./test.csv') #Dataset con nuevos datos para poder probar nuestro nuevo modelo con datos distintos
df_test.head()
```

```python
#Buscar si hay filas duplicadas y eliminarlas
df_test.drop_duplicates()
```

```python
test_id = df_test['id'] #Aislamos los id ya que no los queremos para entrenar
test_without_id = df_test[['fruitset','fruitmass','seeds']];
test_without_id =test_without_id.dropna();
```

```python
#Prediccion del bloque de pruebas
y_pred_regresion_test = modelregresion.predict(pd.DataFrame(test_without_id))
```

```python
y_pred_regresion_test
```

```python
#Creamos nuestro dataframe con las predicciones que hemos hecho junto a su id correspondiente
df_sample = pd.DataFrame({
    'id': test_id,
    'yield': y_pred_regresion_test
})
df_sample.to_csv('sample_submission.csv', index=False)
```
